```python
# Disease Outbreak Analysis

## Introduction
This project aims to track and analyze the spread of a disease outbreak. The primary objectives are to:
- Understand the temporal trends of the outbreak.
- Identify geographic hotspots.
- Visualize the spread and impact of the disease.

## Dataset
We will use a dataset containing information on disease outbreaks, including dates, locations, and the number of cases.
## Data Collection and Preparation
First, we load the dataset and perform initial data preparation.
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv("ebola_2014_2016_clean.csv")

# Display basic information
print(data.info())
print(data.describe())

# Display the first few rows of the dataset
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the trend of cases over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Cumulative no. of confirmed, probable and suspected cases', data=data, hue='Country')
plt.title('Cumulative Number of Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Number of Cases')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Plotting the trend of deaths over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Cumulative no. of confirmed, probable and suspected deaths', data=data, hue='Country')
plt.title('Cumulative Number of Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Number of Deaths')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
## Distribution Analysis, We will explore the distribution of cases and deaths across different countries.

# Distribution of cases by country
plt.figure(figsize=(14, 8))
sns.barplot(x='Country', y='Cumulative no. of confirmed, probable and suspected cases', data=data)
plt.title('Cumulative Number of Cases by Country')
plt.xlabel('Country')
plt.ylabel('Cumulative Number of Cases')
plt.xticks(rotation=90)
plt.show()

# Distribution of deaths by country
plt.figure(figsize=(14, 8))
sns.barplot(x='Country', y='Cumulative no. of confirmed, probable and suspected deaths', data=data)
plt.title('Cumulative Number of Deaths by Country')
plt.xlabel('Country')
plt.ylabel('Cumulative Number of Deaths')
plt.xticks(rotation=90)
plt.show()
```

```python
## Correlation Analysis
## We will explore the correlation between the number of cases and deaths.

import numpy as np

correlation_data = data[['Cumulative no. of confirmed, probable and suspected cases', 
                         'Cumulative no. of confirmed, probable and suspected deaths']]

# Computing the correlation matrix
correlation_matrix = correlation_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
